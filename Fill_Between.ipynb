```python
import matplotlib.pyplot as plt 

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.fill_between(x, y)
plt.plot(x,y,color = "r")
plt.title("Python")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.fill_between(x = [2,4], y1=2, y2=4)
plt.plot(x,y,color = "r")
plt.title("Python")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.fill_between(x = [2,4], y1=2, y2=4 , color = "g")
plt.plot(x,y,color = "r")
plt.title("Python")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
import numpy as np
x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.fill_between(x,y,color = "g" , where =(x>=2) & (x<=4))
plt.plot(x,y,color = "r")
plt.title("Python")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```

```python
plt.fill_between(x,y,color = "g" , where =(x>=2) & (x<=4) , alpha =0.5)
plt.plot(x,y,color = "r")
plt.title("Python")
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.show()
```
